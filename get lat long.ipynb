```python
from bs4 import BeautifulSoup as bs
import requests

res = requests.get("https://en.wikipedia.org/wiki/List_of_constituencies_of_the_West_Bengal_Legislative_Assembly")
soup = bs(res.text, "html.parser")
wiki_links = {}
for link in soup.find_all(class_="wikitable"):
    li = link.find_all("a")
    for i in li:
        url = i.get("href", "")
        wiki_links[i.text.strip()] = url
```

```python
wiki_links['Mekliganj']
```

```python
for k,v in wiki_links.items():
    req = requests.get("https://en.wikipedia.org"+v).text
    soup = bs(req, 'lxml')
    latitude = soup.find("span", {"class": "latitude"})
    longitude = soup.find("span", {"class": "longitude"})
    if latitude:
        print(k, " --- ", latitude.text, longitude.text)
```
